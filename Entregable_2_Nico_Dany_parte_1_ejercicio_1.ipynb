Además, filtramos por zipcode y calculamos el promedio de los precios diarios, semanales y mensuales.

```python
query5 = """
SELECT zipcode,  
    AVG(price) as price_mean, 
    AVG(weekly_price) as weekly_price_mean, 
    AVG(monthly_price) as monthly_price_mean
FROM (
    SELECT 
        CAST(zipcode AS float) as zipcode,
        price,
        weekly_price,
        monthly_price
    FROM melb_data_airbnb
      )
WHERE zipcode IS NOT NULL
GROUP BY zipcode
ORDER BY zipcode DESC
"""
with engine.connect() as con:
      rs = con.execute(query5)
      print(query5)
      for row in rs:
          print(row)
```

Ahora aplicamos esta consulta pero a la unión de las dos tablas

```python
query6 = """
SELECT * FROM melb_data
LEFT JOIN (
  SELECT zipcode ,  
    AVG(price) as price_mean, 
    AVG(weekly_price) as weekly_price_mean, 
    AVG(monthly_price) as monthly_price_mean
  FROM (
    SELECT 
        CAST(zipcode AS float) as zipcode,
        price,
        weekly_price,
        monthly_price
    FROM melb_data_airbnb
        )
  WHERE zipcode IS NOT NULL
  GROUP BY zipcode
  ORDER BY zipcode DESC
          )
      ON melb_data_airbnb.zipcode = melb_data.Postcode
"""
with engine.connect() as con:
      rs = con.execute(query6)
      print(query6)
      for row in rs:
          print(row)
```
